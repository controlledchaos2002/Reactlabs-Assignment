```python
data = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
```

```python
data.head(10)
```

```python
actor_frequency = {}
```

```python
# Iterate through each row of the dataset
for index, row in data.iterrows():
    actors = row['cast']
    
    # Check if the 'cast' field is not empty
    if isinstance(actors, str):
        # Split the actors by comma and remove leading/trailing whitespace
        actor_list = [actor.strip() for actor in actors.split(',')]
        
        # Update the actor frequency dictionary
        for actor in actor_list:
            if actor in actor_frequency:
                actor_frequency[actor] += 1
            else:
                actor_frequency[actor] = 1

# Sort the actor frequencies in descending order
sorted_actor_frequency = sorted(actor_frequency.items(), key=lambda x: x[1], reverse=True)

# Print the top 10 actors and their frequencies
print("Top 10 Actors and Their Frequency of Appearance:")
for actor, frequency in sorted_actor_frequency[:10]:
    print(f"{actor}: {frequency}")
```
